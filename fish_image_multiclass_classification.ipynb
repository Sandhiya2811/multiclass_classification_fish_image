```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Image Preprocessing

```python
img_size = (224, 224)
batch_size = 32

train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

val_gen = ImageDataGenerator(rescale=1./255)
```

## Load Train & Validation Data

```python
train_data = train_gen.flow_from_directory(
    "/content/images.cv_jzk6llhf18tm3k0kyttxz/data/train",
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical"
)

val_data = val_gen.flow_from_directory(
    "/content/images.cv_jzk6llhf18tm3k0kyttxz/data/val",
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical"
)
```

## Build CNN Model

```python
model = Sequential()

# Block 1
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.3))

# Block 2
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.3))

# Block 3
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.4))

# Fully Connected Layer
model.add(Flatten())
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

# Output Layer (11 classes)
model.add(Dense(11, activation='softmax'))
```

## Compile the Model

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

## Train the Model

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=100
)
```

## Test Accuracy

```python
test_gen = ImageDataGenerator(rescale=1./255)

test_data = test_gen.flow_from_directory(
    "/content/images.cv_jzk6llhf18tm3k0kyttxz/data/test",
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical"
)

model.evaluate(test_data)
```

## Predict New Fish Image

```python
from tensorflow.keras.preprocessing import image
import numpy as np

img = image.load_img("/content/images.cv_jzk6llhf18tm3k0kyttxz/data/test/fish sea_food black_sea_sprat/00JL3HEP2KW2.jpg",
                     target_size=(224,224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0

prediction = model.predict(img_array)
class_index = np.argmax(prediction)

print("Predicted class index:", class_index)
print("Fish type:", train_data.class_indices)
```

```python
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np
```

## Prepare Image Generators

```python
img_size = (224, 224)
batch_size = 32

train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

val_gen = ImageDataGenerator(rescale=1./255)

train_data = train_gen.flow_from_directory(
    "/content/images.cv_jzk6llhf18tm3k0kyttxz/data/train",
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical"
)

val_data = val_gen.flow_from_directory(
    "/content/images.cv_jzk6llhf18tm3k0kyttxz/data/val",
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical"
)

test_gen = ImageDataGenerator(rescale=1./255)
test_data = test_gen.flow_from_directory(
    "/content/images.cv_jzk6llhf18tm3k0kyttxz/data/test",
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

# VGG16

```python
# load  pretrained model

base_model_vgg = VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)


base_model_vgg.trainable = False

# add layers
x = base_model_vgg.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation="relu")(x)
x = Dropout(0.5)(x)
output = Dense(11, activation="softmax")(x)


# model = Sequential()
# model.add(base_model_vgg)
# model.add(GlobalAveragePooling2D())
# model.add(Dense(128, activation='relu'))
# model.add(Dropout(0.5))
# model.add(Dense(11, activation='softmax'))


model_vgg = Model(inputs=base_model_vgg.input, outputs=output)

# compile model
model_vgg.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# train model
history = model_vgg.fit(
    train_data,
    validation_data=val_data,
    epochs=25
)


```

## evaluate the model

```python
test_loss, test_acc = model_vgg.evaluate(test_data)
print("Fine-Tuned vgg16 Test Accuracy:", test_acc*100)
```

## predict new fish
